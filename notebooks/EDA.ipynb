# Section 4 — **Customer Overview**

Before analyzing the factors that contribute to customer churn, it is important to understand the overall customer population.

This section provides a high-level overview of the customer base by examining the total number of customers, active and churned customers, and the overall churn rate.

These key performance indicators (KPIs) establish the baseline of the dataset and provide business stakeholders with a clear understanding of customer retention performance before conducting more detailed analyses.

The objective of this section is to answer the following business questions:

- How many customers are included in the dataset?
- How many customers have churned?
- How many customers remain active?
- What is the overall customer churn rate?

The results from this section will serve as the foundation for all subsequent analyses throughout this notebook.

```python
# Customer Overview KPIs

total_customers = len(df)

active_customers = (df["Churn"] == "No").sum()

churned_customers = (df["Churn"] == "Yes").sum()

churn_rate = (churned_customers / total_customers) * 100

print("CUSTOMER OVERVIEW")

print(f"Total Customers     : {total_customers:,}")
print(f"Active Customers    : {active_customers:,}")
print(f"Churned Customers   : {churned_customers:,}")
print(f"Customer Churn Rate : {churn_rate:.2f}%")
```

```python
# KPI Summary

kpi_summary = pd.DataFrame({

    "Metric":[
        "Total Customers",
        "Active Customers",
        "Churned Customers",
        "Customer Churn Rate (%)"
    ],

    "Value":[
        total_customers,
        active_customers,
        churned_customers,
        round(churn_rate,2)
    ]

})

kpi_summary
```

```python
# Customer Status Distribution


plt.figure(figsize=(8,6))

sns.countplot(
    data=df,
    x="Churn"
)

plt.title("Customer Status Distribution", fontsize=16, weight="bold")
plt.xlabel("Customer Status")
plt.ylabel("Number of Customers")

plt.show()
```

```python
# Customer Churn Distribution

plt.figure(figsize=(7,7))

df["Churn"].value_counts().plot(

    kind="pie",

    autopct="%1.1f%%",

    startangle=90

)

plt.ylabel("")

plt.title("Customer Churn Distribution", fontsize=16, weight="bold")

plt.show()
```

```python
# Customer Churn Percentage

churn_percentage = (

    df["Churn"]

    .value_counts(normalize=True)

    .mul(100)

    .round(2)

)

churn_percentage
```

# 5.1 **Gender Analysis**


Gender is one of the most fundamental demographic characteristics available within the customer dataset.

Although gender alone may not directly determine customer churn, understanding whether churn behavior differs between male and female customers provides valuable business insight and establishes an important baseline for demographic analysis.

This section explores both the overall gender distribution and the relationship between gender and customer churn.

The objective is to answer the following business question:

**Does customer gender significantly influence customer churn?**

```python
# Gender Distribution

gender_summary = df["gender"].value_counts()

gender_percentage = round(
    df["gender"].value_counts(normalize=True)*100,
    2
)

gender_table = pd.DataFrame({

    "Customers": gender_summary,

    "Percentage (%)": gender_percentage

})

gender_table
```

```python
# Gender Distribution

plt.figure(figsize=(8,6))

sns.countplot(
    data=df,
    x="gender"
)

plt.title(
    "Customer Gender Distribution",
    fontsize=16,
    weight="bold"
)

plt.xlabel("Gender")

plt.ylabel("Number of Customers")

plt.show()
```

### **Business Interpretation**

The gender distribution provides an overview of the customer population.

A balanced gender distribution indicates that the customer base is relatively representative and minimizes demographic bias during subsequent analyses.

If one gender substantially dominates the dataset, additional investigation may be required to determine whether marketing strategies or product offerings disproportionately attract a specific customer group.

```python
# Gender vs Customer Churn


gender_churn = pd.crosstab(

    df["gender"],

    df["Churn"]

)

gender_churn
```

```python
# Gender vs Customer Churn


plt.figure(figsize=(8,6))

sns.countplot(

    data=df,

    x="gender",

    hue="Churn"

)

plt.title(

    "Customer Churn by Gender",

    fontsize=16,

    weight="bold"

)

plt.xlabel("Gender")

plt.ylabel("Number of Customers")

plt.legend(title="Churn")

plt.show()
```

```python
# Churn Rate by Gender

gender_churn_rate = round(

    pd.crosstab(

        df["gender"],

        df["Churn"],

        normalize="index"

    )*100,

    2

)

gender_churn_rate
```

```python
# Customer Churn Rate by Gender


gender_churn_rate["Yes"].plot(

    kind="bar",

    figsize=(8,6)

)

plt.title(

    "Customer Churn Rate by Gender",

    fontsize=16,

    weight="bold"

)

plt.ylabel("Churn Rate (%)")

plt.xlabel("Gender")

plt.show()
```

## **Business Interpretation**


The gender-based churn analysis evaluates whether customer churn differs between male and female customers.

If both genders exhibit similar churn rates, gender is unlikely to be a significant predictor of customer churn and may have limited influence on customer retention strategies.

Conversely, a substantial difference in churn rates would suggest that gender-specific engagement strategies could improve customer retention.

The observed results provide valuable insight into whether demographic segmentation by gender should be considered during predictive modeling and customer retention planning.

```python
# Visualization Helper Functions


def style_chart(title, xlabel, ylabel):

    plt.title(
        title,
        fontsize=18,
        weight="bold",
        pad=15
    )

    plt.xlabel(xlabel, fontsize=12)
    plt.ylabel(ylabel, fontsize=12)

    plt.grid(
        axis="y",
        linestyle="--",
        alpha=.3
    )

    plt.tight_layout()


def add_value_labels(ax):

    for container in ax.containers:

        ax.bar_label(
            container,
            padding=3,
            fontsize=10
        )

print("Visualization helper functions loaded.")
```

```python
# Project Color Palette


PRIMARY   = "#6366F1"   # Indigo
SECONDARY = "#06B6D4"   # Cyan
SUCCESS   = "#22C55E"   # Green
WARNING   = "#F59E0B"   # Orange
DANGER    = "#EF4444"   # Red
PURPLE    = "#8B5CF6"
PINK       = "#EC4899"
GRAY      = "#6B7280"

project_palette = [
    PRIMARY,
    SECONDARY,
    SUCCESS,
    WARNING,
    DANGER,
    PURPLE,
    PINK
]

# Churn Colors
churn_colors = {
    "No": SUCCESS,
    "Yes": DANGER
}

print("Color palette loaded successfully.")
```

```python
# Customer Service Variables

service_columns = [
    "PhoneService",
    "MultipleLines",
    "InternetService",
    "OnlineSecurity",
    "OnlineBackup",
    "DeviceProtection",
    "TechSupport",
    "StreamingTV",
    "StreamingMovies"
]

print("Customer service variables selected:")
for column in service_columns:
    print(f"- {column}")
```

```python
# Reusable Customer Service Analysis Function


def analyze_service(column, display_name):
    """
    Analyze the distribution and churn rate of a service variable.
    """

    # Service summary
    summary = pd.DataFrame({
        "Customers": df[column].value_counts(),
        "Percentage (%)": (
            df[column]
            .value_counts(normalize=True)
            .mul(100)
            .round(2)
        )
    })

    display(summary)

    # Distribution chart
    plt.figure(figsize=(10, 6))

    ax = sns.countplot(
        data=df,
        x=column,
        hue=column,
        palette=project_palette,
        legend=False
    )

    style_chart(
        f"Customer Distribution by {display_name}",
        display_name,
        "Number of Customers"
    )

    add_value_labels(ax)
    plt.show()

    # Churn count chart
    plt.figure(figsize=(10, 6))

    ax = sns.countplot(
        data=df,
        x=column,
        hue="Churn",
        palette=churn_colors
    )

    style_chart(
        f"Customer Churn by {display_name}",
        display_name,
        "Number of Customers"
    )

    add_value_labels(ax)
    plt.legend(title="Churn")
    plt.show()

    # Churn-rate table
    churn_rate = (
        pd.crosstab(
            df[column],
            df["Churn"],
            normalize="index"
        )
        .mul(100)
        .round(2)
    )

    display(churn_rate)

    # Churn-rate chart
    churn_plot = churn_rate["Yes"].sort_values(ascending=False)

    plt.figure(figsize=(10, 6))

    ax = sns.barplot(
        x=churn_plot.index,
        y=churn_plot.values,
        hue=churn_plot.index,
        palette=project_palette,
        legend=False
    )

    style_chart(
        f"Customer Churn Rate by {display_name}",
        display_name,
        "Churn Rate (%)"
    )

    for container in ax.containers:
        ax.bar_label(
            container,
            fmt="%.1f%%",
            padding=4,
            fontsize=10
        )

    plt.ylim(0, churn_plot.max() * 1.20)
    plt.show()

    return churn_rate
```
