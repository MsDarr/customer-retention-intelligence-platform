## Duplicate Record Assessment

Duplicate observations may introduce bias into predictive models and distort statistical analysis.

This section identifies duplicate customer records and removes them if necessary.

```python
duplicate_count = df.duplicated().sum()

print(f"Duplicate Records Found: {duplicate_count}")

if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate records removed.")
else:
    print("No duplicate records found.")
```

## Missing Value Assessment

Missing values reduce data quality and may negatively impact predictive performance. This section identifies incomplete observations before applying appropriate treatment strategies.

```python
df.isnull().sum()
```

## Data Type Validation

Incorrect data types may lead to inaccurate statistical calculations and machine learning errors.

This section validates and corrects the data types where necessary.

```python
df.dtypes
```

```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)
```

```python
df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

print(df.shape)
```

## Feature Selection

The customerID column is a unique identifier used only for customer tracking.

Since it contains no predictive information, it is removed to prevent unnecessary complexity during exploratory analysis and machine learning.

```python
if "customerID" in df.columns:
    df.drop(columns=["customerID"], inplace=True)
    print("customerID removed successfully.")
else:
    print("customerID already removed.")
```
